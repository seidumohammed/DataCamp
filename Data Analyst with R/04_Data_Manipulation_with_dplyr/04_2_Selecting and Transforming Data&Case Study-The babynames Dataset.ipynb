# 3. Selecting and Transforming Data
Learn advanced methods to select and transform columns. Also learn about select helpers, which are functions that specify criteria for columns you want to choose, as well as the rename and transmute verbs.

````r
library(dplyr)
````

````r
counties <- readRDS("counties.rds")
````

## Selecting
### Select a range

````r
counties %>%
    select(state, county, drive:work_at_home) %>%
    arrange(drive)
````

The cities where the fewest people drive to work are a few regions in Alaska and in New York City. Other interesting insights are in New York people mostly take transit to work, while in the small parts of Alaska they mostly walk.

### Contains


````r
counties %>%
    select(state, county, contains("work"))
````

### Starts with

````r
counties %>%
    select(state, county, starts_with("income"))
````

### Other helpers
- `contains()`
- `starts_with()`
- `ends_with()`
- `last_col()`

### Removing a variable
If decide that the census id for each county is not very helpful, use `select(-census_id)` to get rid of just that column.

````r
counties %>%
    select(-census_id)
````

````r
counties %>%
  # Select state, county, population, and industry-related columns
  select(state, county, population, professional:production) %>%
  # Arrange service in descending order 
  arrange(desc(service))
````

## The rename verb


````r
counties_selected <- counties %>%
    select(state, county, population, unemployment)
````

### Rename a column

````r
counties_selected %>%
    rename(unemployment_rate = unemployment) %>%
    top_n(10, unemployment_rate)
````

### Combine verbs

````r
counties_selected %>%
    select(state, county, population, unemployment_rate = unemployment) %>%
    top_n(10, unemployment_rate)
````

## The transmute verb
- Transmute is like a combination of select and mutate
- Returns a subset of columns that are transformed and changed

### Select and calculate

````r
counties %>%
    transmute(state, county, fraction_men = men / population)
````

### Summary

_ | Keeps only specified variables | Keeps other variables
:---|:---|:---
Can't change values | `select()` | `rename()`
Can change values | `transmute()` | `mutate()`

## Exercise

````r
# Using transmute

counties %>%
  # Keep the state, county, and populations columns, and add a density column
  transmute(state, county, population, density = population / land_area) %>%
  # Filter for counties with a population greater than one million 
  filter(population > 1000000) %>%
  # Sort density in ascending order 
  arrange(density)
````

## Choosing among the four verbs

- Choose the right verb for changing the name of the `unemployment` column to `unemployment_rate`

```r
# Change the name of the unemployment column
counties %>%
    rename(unemployment_rate = unemployment)
```



- Choose the right verb for keeping only the columns `state`, `county`, and the ones containing poverty.

```r
# Keep the state and county columns, and the columns containing poverty
counties %>%
    select(state, county, 
           contains("poverty"))
```



- Calculate a new column called `fraction_women` with the fraction of the population made up of women, without dropping any columns.

```r
# Calculate the fraction_women column without dropping the other columns
counties %>%
    mutate(fraction_women = women / population)
```



- Keep only three columns: the `state`, `county`, and `employed / population`, which you'll call `employment_rate`.

```r
# Keep only the state, county, and employment_rate columns
counties %>%
    transmute(state, county, 
              employment_rate = employed / population)
```

## The babynames data
In the babynames table, each observation, or row, represents one combination of a year and a name, like 1880 and John. The third variable, number, represents the number of babies born in the United States with that name. For example, in 1880 there were 102 babies born named Aaron.

## Grouped mutates
### babynames graph
In one of the last exercises, the graph was created. The y-axis represents the number of people born in each year with the name Matthew, Steven, or Thomas. 

But remember that a different total number of babies are born in each year, and what we're interested in isn't exactly the number who are born. Rather, we're interested in what percentage of people born in that year have that name. To calculate that, you'll need to work with the total number of people born in each year. 

### group_by() and summarize()

````r
babynames %>%
    group_by(year) %>%
    summarize(year_total = sum(number))
````

### Combining group_by() and mutate()

````r
babynames %>%
    group_by(year) %>%
    mutate(year_total = sum(number))
````

### ungroup() & Add the fraction column

````r
babynames_with_fraction <- babynames %>%
    group_by(year) %>%
    mutate(year_total = sum(number)) %>%
    ungroup() %>%
    mutate(fraction = number / year_total * 100)

babynames_with_fraction
````

````r
selected_names_with_fraction <- babynames_with_fraction %>%
  # Filter for the names Steven, Thomas, and Matthew 
  filter(name %in% c("Steven", "Thomas", "Matthew"))

# Plot the names using a different color for each name
ggplot(selected_names_with_fraction, aes(x = year, y = fraction, color = name)) +
  geom_line()
````

## Exercise

````r
# Finding the year each name is most common

# Calculate the fraction of people born each year with the same name
babynames %>%
  group_by(year) %>%
  mutate(year_total = sum(number)) %>%
  ungroup() %>%
  mutate(fraction = number / year_total) %>%
  # Find the year each name is most common
  group_by(name) %>%
  top_n(1, fraction)
````

````r
# Adding the total and maximum for each name

babynames %>%
  # Add columns name_total and name_max for each name
  group_by(name) %>%
  mutate(name_total = sum(number),
         name_max = max(number)) %>%
  # Ungroup the table 
  ungroup() %>%
  # Add the fraction_max column containing the number by the name maximum 
  mutate(fraction_max = number / name_max)
````

*This tells that the name Abe was at 18.5% of its peak in the year 1880.*

````r
# Visualizing the normalized change in popularity

names_normalized <- babynames %>%
    group_by(name) %>%
    mutate(name_total = sum(number), 
           name_max = max(number)) %>%
    ungroup() %>%
    mutate(fraction_max = number / name_max)

names_filtered <- names_normalized %>%
  # Filter for the names Steven, Thomas, and Matthew
  filter(name %in% c("Steven", "Thomas", "Matthew"))

# Visualize these names over time
ggplot(names_filtered, aes(x = year, y = fraction_max, color = name)) +
  geom_line()
````

*The line for each name hits a peak at 1, although the peak year differs for each name.*

## Window functions
You've made some visualizations, and discovered a few names that have gone through major changes over time. But what if you want to look at the biggest changes within each name? To do this, you'd have to find differences between each pair of consecutive years.

### Window function
A window function takes a vector, and returns another vector of the same length. 

- `lag()` function

For example, suppose we had a vector of 1, 3, 6, and 14. We can lag this vector, which means moving each item to the right by one. 

````r
v <- c(1, 3, 6, 14)
v
````

````r
lag(v)
````

Now the first item is `NA`, meaning it's missing, but it's followed by 1, 3, and 6: the item just prior to it in the original vector.

###  Compare consecutive steps
Now, why is this useful? Because by lining up each item in the vector with the item directly before it, we can compare consecutive steps and calculate the changes. 

With `v - lag(v)`, we're asking "What is each value once we've subtracted the previous one?"

````r
v - lag(v)
````

### Changes in popularity of a name
Now that we know how to calculate the difference between consecutive values in a vector, we can use that in a grouped mutate to find the changes in the popularity of one name in consecutive years. 

````r
babynames_fraction <- babynames %>%
    group_by(year) %>%
    mutate(year_total = sum(number)) %>%
    ungroup() %>%
    mutate(fraction = number / year_total)
````

### Matthew
Let's start by picking a name, like Matthew, and filtering for it. Then arrange in ascending order.

````r
babynames_fraction %>%
    filter(name == "Matthew") %>%
    arrange(year)
````

Notice that the fraction of babies named Matthew started around .0005, then went down to .004, then back up a bit.

### Matthew over time
To quantify that, you could use a mutate with the lag window function. We want to take each fraction, and subtract the "lagged" fraction, with `fraction - lag(fraction)`. 

````r
babynames_fraction %>%
    filter(name == "Matthew") %>%
    arrange(year) %>%
    mutate(difference = fraction - lag(fraction))
````

Notice that the first observation is missing a difference, because there is no previous year. After that, we can see whether Matthew went up or down within each year.

### Biggest jump in popularity
What if we wanted to know the biggest jump that the name Matthew took in popularity?

````r
babynames_fraction %>%
    filter(name == "Matthew") %>%
    arrange(year) %>%
    mutate(difference = fraction - lag(fraction)) %>%
    arrange(desc(difference))
````

We could sort in descending order of the difference column, and see that the biggest jumps, when Matthew got much more popular, were in 1975 and 1970.

### Changes within every name
Now what if instead of looking at one name, you wanted to look at the changes within every name? You'll need to do this as a grouped mutate. This ensures you won't include differences between one name and a different one.

````r
babynames_fraction %>%
    arrange(name, year) %>%
    mutate(difference = fraction - lag(fraction)) %>%
    group_by(name) %>%
    arrange(desc(difference))
````

````r
# Using ratios to describe the frequency of a name

babynames_fraction %>%
  # Arrange the data in order of name, then year 
  arrange(name, year) %>%
  # Group the data by name
  group_by(name) %>%
  # Add a ratio column that contains the ratio of fraction between each year 
  mutate(ratio = fraction / lag(fraction))
````

*Notice that the first observation for each name is missing a ratio, since there is no previous year.*

````r
# Biggest jumps in a name

babynames_ratios_filtered <- babynames_fraction %>%
  arrange(name, year) %>%
  group_by(name) %>%
  mutate(ratio = fraction / lag(fraction)) %>%
  filter(fraction >= 0.00001)

babynames_ratios_filtered %>%
  # Extract the largest ratio from each name 
  top_n(1, ratio) %>%
  # Sort the ratio column in descending order 
  arrange(desc(ratio)) %>%
  # Filter for fractions greater than or equal to 0.001
  filter(fraction >= 0.001)
````

*Some of these can be interpreted: for example, Grover Cleveland was a president elected in 1884.*
