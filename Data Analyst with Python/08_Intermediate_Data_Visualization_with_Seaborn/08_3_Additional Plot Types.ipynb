# 3. Additional Plot Types
**Overview of more complex plot types included in Seaborn.**

````python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
````

## Categorical Plot Types

### Categorical Data
- Data which takes on a limited and fixed number of values
- Normally combined with numeric data
- Examples include:
    - Geography (country, state, region)
    - Gender
    - Ethnicity
    - Blood type
    - Eye color

### Plot types
Seaborn breaks categorical data plots into three groups. 
1. The first group includes the `stripplot()` and `swarmplot()`, which **show all the individual observations** on the plot.
2. The second category contains the familiar `boxplot()`, as well as the `violinplot()` and `lvplot()`(before v0.9.0; it's been renamed to `boxenplot()`). These plots show an **abstract representation** of the categorical data.
3. The final group of plots show **statistical estimates** of the categorical variables. The `barplot()` and `pointplot()` contain useful summaries of data. The `countplot()` shows the number of instances of each observation.

#### 1-1. Plots of each observation - stripplot
Seaborn's `stripplot()` **shows every observation** in the dataset. In some cases, it can be difficult to see individual data points.

- Example:

```python
sns.stripplot(data=df, y='column_1', 
              x='column_2', jitter=True)
```

#### 1-2. Plots of each observation - swarmplot
We can plot a more sophisticated visualization of all the data using a `swarmplot()`. This plot uses a complex algorithm to **place the observations in a manner where they do not overlap**. 

The downside of this approach is that the `swarmplot()` does not scale well to large datasets.

- Example:

```python
sns.swarmplot(data=df, y='column_1', x='column_2')
```

#### 2-1. Abstract representations - boxplot
The next category of plots **show abstract representations** of the data. A `boxplot()` is the most common of this type. This plot is used to show several measures related to the distribution of data, including the median, upper and lower quartiles, as well as outliers.

- Example:

```python
sns.boxplot(data=df, y='column_1', x='column_2')
```

#### 2-2. Abstract representation - violinplot
The `violinplot()` is a **combination of a kernel density plot and a box plot** and can be suitable for **providing an alternative view of the distribution of data**. Because the plot uses a kernel density calculation it does not show all data points. 

This can be useful for displaying large datasets but it can be computationally intensive to create.

- Example:

```python
sns.vilolinplot(data=df, y='column_1', x='column_2')
```

#### 2-3. Abstract representation - lvplot (before v0.9.0; it's been renamed to boxenplot)
The final plot in this grouping is the `lvplot()`(before v0.9.0; it's been renamed to `boxenplot()`), which stands for Letter Value plot. The API is the same as the `boxplot()` and `violinplot()` but **can scale more effectively to large datasets**.

The `lvplot()`(before v0.9.0; it's been renamed to `boxenplot()`) is a hybrid between a `boxplot()` and `violinplot()` and is relatively quick to render and easy to interpret.

- Example:

```python
sns.lvplot(data=df, y='column_1', x='column_2')
```

#### 3-1. Statistical estimates - barplot
The final category of plots are **statistical estimates** of the data. The `barplot()` **shows an estimate of the value *as well as* a confidence interval**. 

- Example:

```python
sns.barplot(data=df, y='column_1', 
            x='column_2', hue='column_3')
```

#### 3-2. Statistical estimates - pointplot
The `pointplot()` is similar to the `barplot()` in that it **shows a summary measure and confidence interval**. A `pointplot()` can be very useful for **observing how values change across categorical values**.

- Example:
```python
sns.pointplot(data=df, y='column_1', 
              x='column_2', hue='column_3')
```

#### 3-3. Statistical estimates - countplot
The final categorical plot is the `countplot()`, which **displays the number of instances of each variable**.

- Example:
```python
sns.countplot(data=df, y='column_1', x='column_2')
```

## `stripplot()` and `swarmplot()`
Many datasets have categorical data and Seaborn supports several useful plot types for this data. In this example, we will continue to look at the 2010 School Improvement data and segment the data by the types of school improvement models used.

````python
grants = pd.read_csv('schoolimprovement2010grants.csv')
````

Here is the KDE distribution of the Award Amounts:

````python
grants['Award_Amount'].plot.kde()
plt.show()
````

While this plot is useful, there is a lot more we can learn by looking at the individual `Award_Amounts` and how they are distributed among the 4 categories.

- Create a `stripplot` of the `Award_Amount` with the `Model Selected` on the y axis with `jitter` enabled.

````python
# Create the stripplot
sns.stripplot(data=grants, x='Award_Amount', y='Model Selected', jitter=True)

plt.show()
````

- Create a `swarmplot()` of the same data, but also include the `hue` by `Region`.

````python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=grants, x='Award_Amount', y='Model Selected', hue='Region', size=1.88)

plt.show()
````

*Categorical plots that show each observation can be very powerful tools for gaining insight into your data.*

## boxplots, violinplots and lvplots(before v0.9.0; it's been renamed to boxenplot)
Seaborn's categorical plots also support several abstract representations of data. The API for each of these is the same so it is very convenient to try each plot and see if the data lends itself to one over the other.

In this exercise, we will use the color palette options.

- Create and display a `boxplot` of the data with `Award_Amount` on the x axis and `Model Selected` on the y axis.

````python
# Create a boxplot
sns.boxplot(data=grants, x='Award_Amount', y='Model Selected')

plt.show()
plt.clf()
````

- Create and display a similar `violinplot` of the data, but use the `husl` palette for colors.

````python
# Create a violinplot with the husl palette
sns.violinplot(data=grants, x='Award_Amount', y='Model Selected', palette='husl')

plt.show()
plt.clf()
````

- Create and display an `lvplot`(before v0.9.0; it's been renamed to `boxenplot()`) using the `Paired` palette and the `Region` column as the `hue`.

````python
# Create a lvplot(before v0.9.0; it's been renamed to `boxenplot()`) 
# with the Paired palette and the Region column as the hue
sns.boxenplot(data=grants, x='Award_Amount', y='Model Selected', palette='Paired', hue='Region')

plt.show()
plt.clf()
````

## barplots, pointplots and countplots
The final group of categorical plots are `barplots`, `pointplots` and `countplot` which create statistical summaries of the data. The plots follow a similar API as the other plots and allow further customization for the specific problem at hand.

- Create a `countplot` with the dataframe and `Model Selected` on the y axis and the color varying by `Region`.

````python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=grants, y='Model Selected', hue='Region')

plt.show()
plt.clf()
````

- Create a `pointplot` with the dataframe and `Model Selected` on the x-axis and `Award_Amount` on the y-axis.
- Use a `capsize` in the `pointplot` in order to add caps to the error bars.

````python
# Create a pointplot and include the capsize in order to show caps on the error bars
sns.pointplot(data=grants, y='Award_Amount', x='Model Selected', capsize=0.1)

plt.show()
plt.clf()
````

- Create a `barplot` with the same data on the x and y axis and change the color of each bar based on the `Region` column.

````python
# Create a barplot with each Region shown as a different color
sns.barplot(data=grants, y='Award_Amount', x='Model Selected', hue='Region')

plt.show()
plt.clf()
````

*The pointplot and barplot can be useful visualizations for understanding the variations of categorical data.*

## Regression Plots
Since regression plots are a very important tool in the data scientist's toolbox, Seaborn has a robust API that supports sophisticated analysis of datasets. We will spend some time looking at regression analysis of bicycle-sharing data from Washington, DC.

````python
bike_df = pd.read_csv('bike_share.csv')
````

### Bicycle Dataset
- Aggregated bicycle sharing data in Washington DC
- Data includes:
    - Rental amounts
    - Weather information
    - Calendar information
- Can we predict rental amounts?

We can use Seaborn's regression plotting tools to evaluate the data from multiple perspectives and look for relationships between these numeric variables.

### Plotting with `regplot()`
Like most of the Seaborn functions, the function requires the definition of the data and the x and y variables. In this example, we include a unique marker for the observations. 

````python
sns.regplot(data=bike_df, x='temp', y='total_rentals', marker='+')
plt.show()
plt.clf()
````

Upon first glance, it looks like there is a relationship between temperature and total bike rentals in a day. This intuitively makes sense. People like to bike when the weather is warm.

### Evaluating regression with `residplot()`
- A residual plot is useful for evaluating the fit of a model
- Seaborn supports through `residplot` function

Ideally, the residual values in the plot should be plotted randomly across the horizontal line. 

````python
sns.residplot(data=bike_df, x='temp', y='total_rentals')
plt.show()
plt.clf()
````

In this specific example, the data looks like it might have a slight curve suggesting a nonlinear model might be appropriate.

### Polynomial regression
- Seaborn supports polynomial regression using the `order` parameter

If a value greater than 1 is passed to the order parameter of regplot(), then Seaborn will attempt a polynomial fit using underlying NumPy functions.

````python
sns.regplot(data=bike_df, x='temp', y='total_rentals', order=2)
plt.show()
plt.clf()
````

In this example, Seaborn computes a second order polynomial function for the relationship between temperature and rentals. In this view, it looks like rentals might start to trail off if the weather gets too warm.

### residplot with polynomial regression
The residual plot can interpret the second order polynomial and plot the residual values. 

````python
sns.residplot(data=bike_df, x='temp', y='total_rentals', order=2)
plt.show()
plt.clf()
````

In this example, the values are more randomly distributed, so a second order equation is likely more appropriate for this problem.

### Categorical values
Seaborn also supports regression plots with categorical variables. It might be interesting to see how rentals change over the various months.

````python
sns.regplot(data=bike_df, x='mnth', y='total_rentals', x_jitter=.1, order=2)
plt.show()
plt.clf()
````

In this example, using the jitter parameter makes it easier to see the individual distribution of the rental values for each month.

### Estimators
- In some cases, even with the jitter, it may be difficult to see if there are any trends based on the value of the variable
- Using an estimator for the x value (`X_estimator`) can be useful for highlighting trends

````python
sns.regplot(data=bike_df, x='mnth', y='total_rentals', x_estimator=np.mean, order=2)
plt.show()
plt.clf()
````

This simplified view shows a trend consistent with the seasons in Washington, DC.

## Binning the data
- `x_bins` can be used to divide the data into discrete bins
- The regression line is still fit against all the data

In this case, we can divide the temperature into four bins and Seaborn will take care of calculating those bins for us and plotting the results.

````python
sns.regplot(data=bike_df, x='temp', y='total_rentals', x_bins=4)
plt.show()
plt.clf()
````

This is much quicker than trying to use pandas or some other mechanism to create the bins. This shortcut function can help with getting a quick read on continuous data such as temperature.

## Regression and residual plots
Linear regression is a useful tool for understanding the relationship between numerical variables. Seaborn has simple but powerful tools for examining these relationships.

For these exercises, we will look at some details from the US Department of Education on 4 year college tuition information and see if there are any interesting insights into which variables might help predict tuition costs.

````python
college_df = pd.read_csv('college_datav3.csv')
````

- Plot a regression plot comparing `Tuition` and average SAT scores(`SAT_AVG_ALL`).
- Make sure the values are shown as green triangles.

## Matrix plots
The heatmap is the most common type of matrix plot and can be easily created by Seaborn. These types of matrix plots can be useful for quickly seeing trends in a dataset. 

### Getting data in the right format
- Seaborn's `heatmap()` function requires data to be in a grid format, a matrix
- pandas `crosstab()` is frequently used to manipulate the data

To illustrate this process, we can look at the bicycle data to see how rentals vary by day of the week and month. In this example, we also use the `aggfunc` argument to get the average number of rentals for each of the day and month combinations. 

````python
pd.crosstab(bike_df['mnth'], bike_df['weekday'], 
            values=bike_df['total_rentals'], aggfunc='mean').round(0)
````

This is a very useful capability available in pandas and makes it easy to plot a heatmap once it is in this format.

### Build a heatmap
The heatmap translates the numerical values in the matrix into a color coded grid. In this example, the colors get lighter as the numbers increase.

````python
sns.heatmap(pd.crosstab(bike_df['mnth'], bike_df['weekday'], 
                        values=bike_df['total_rentals'], aggfunc='mean'))
````

Using this approach, it is easy to see that the busiest days are Saturdays in June and September. The color code also illustrates that rental rates are higher in the warmer weather months from May through October.

### Customize a heatmap
The display of the heatmap can be customized in multiple ways to present the most information as clearly as possible. This example illustrates a number of customizations. 
- First, we use `annot` equals `True` to turn on annotations in the individual cells. The `fmt` option ensures that the results are displayed as integers. 
- Next, we use a custom `cmap` of Yellow Green Blue to change the shading we use. By setting `cbar` equals `False`, the color bar is not displayed. 
- Finally, passing a variable to `linewidths` puts some small spacing between the cells so that the values are simpler to view.

````python
sns.heatmap(pd.crosstab(bike_df['mnth'], bike_df['weekday'], 
                        values=bike_df['total_rentals'], aggfunc='mean'), 
            annot=True, fmt='.0f', 
            cmap='YlGnBu', cbar=False, linewidths=.5)
````

This example shows how changing a few parameters in Seaborn really changes the output of the plot.

### Centering a heatmap
- Seaborn support centering the heatmap colors on a dpecific value

The final customization we will look at is centering the `heatmap()` color scheme.
In this example, we center the colormap at the value stored for Sept and Saturday of 9 and 6 using the numeric indices for these values. The effect of this change is that overall color scheme is shifted towards yellows instead of blues.

````python
df_crosstab = pd.crosstab(bike_df['mnth'], bike_df['weekday'], 
                          values=bike_df['total_rentals'], aggfunc='mean')

sns.heatmap(df_crosstab, annot=True, fmt='.0f', 
            cmap='YlGnBu', cbar=True, center=df_crosstab.loc[9, 6])
````

### Plotting a correlation matrix
One common usage for a heatmap is to visually represent the correlation between variables.

- Pandas `corr` function calculates correlations between between the values in the columns
- The output can be converted to a heatmap with seaborn

This visualization can be useful to understand which variables you might want to further study using `regplot()`.

````python
sns.heatmap(bike_df.corr())
````

This plot shows that **the total rentals are most highly correlated with the temp variables**.

## Creating heatmaps
A heatmap is a common matrix plot that can be used to graphically summarize the relationship between two variables. For this exercise, we will start by looking at guests of the Daily Show from 1999 - 2015 and see how the occupations of the guests have changed over time.

The data includes the date of each guest appearance as well as their occupation. For the first exercise, we need to get the data into the right format for Seaborn's `heatmap` function to correctly plot the data. 

````python
guests = pd.read_csv('daily_show_guests_cleaned.csv')
````

- Use pandas' `crosstab()` function to build a table of visits by `Group` and `Year`.
- Print the `pd_crosstab` DataFrame.

````python
# Create a crosstab table of the data
pd_crosstab = pd.crosstab(guests['Group'], guests['YEAR'])
pd_crosstab
````

- Plot the data using Seaborn's `heatmap()`.

````python
# Plot a heatmap of the table
sns.heatmap(pd_crosstab)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
````

## Customizing heatmaps
Seaborn supports several types of additional customizations to improve the output of a heatmap. For this exercise, we will continue to use the Daily Show data.

- Create a crosstab table of `Group` and `YEAR`
- Create a heatmap of the data using the `BuGn` palette
- Disable the `cbar` and increase the `linewidth` to 0.3

````python
# Create the crosstab DataFrame
pd_crosstab = pd.crosstab(guests['Group'], guests['YEAR'])

# Plot a heatmap of the table with no color bar and using the BuGn palette
sns.heatmap(pd_crosstab, cbar=False, cmap='BuGn', linewidths=.3)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

# Show the plot
plt.show()
plt.clf()
````
